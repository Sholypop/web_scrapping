```python
# writing a function to get the playlist

def get_playlist_tracks(username, playlist_id):
    results = sp.user_playlist_tracks(username,playlist_id,market="GE")
    tracks = results['items']
    while results['next']:
        results = sp.next(results)
        tracks.extend(results['items'])
    return tracks
```

```python
tracks=get_playlist_tracks("spotify", "37i9dQZF1DXbwoaqxaoAVr")
```

```python
# compiling the list of songs into a variable
```

```python
list_of_audio_features=[]
for item in range(len(tracks)):
    #print (tracks[item]["track"]["id"])
    list_of_audio_features.append(sp.audio_features(tracks[item]["track"]["id"])[0])
```

```python
# Building a Data frame to store my features
```

```python
df=pd.DataFrame(list_of_audio_features)    
df=df[["danceability","energy","loudness","speechiness","acousticness",
    "instrumentalness","liveness","valence","tempo","id","duration_ms"]]

df
```
