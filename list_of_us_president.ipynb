```python
#importing library

import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#find url and store it in a variable
url = "https://en.wikipedia.org/wiki/List_of_presidents_of_the_United_States"

# get request
response = requests.get(url)
response.status_code # 200 status code means OK!
```

### create soup

```python
# create the 'soup'
soup = BeautifulSoup(response.text, "lxml")

soup
```

```python
soup.select('#mw-content-text > div.mw-parser-output > table > tbody > tr:nth-child(2) > td:nth-child(3) > b > a')
```

```python
soup.select('#mw-content-text > div.mw-parser-output > table > tbody > tr > td > b')[0]
```

### scrapping the list of all president

```python
# scrapping the list of all president

president = []
for i in soup.select('table > tbody > tr > td > b > a'):
    president.append(i)
```

### Scrape the Wikipedia page of each president.

```python
# To make it more "human", we can randomize the waiting time:
from time import sleep
from random import randint
```

```python
# find url and store it in a variable

presi_soups = []

for presi in president:
    # send request
    url = "https://en.wikipedia.org/" + presi["href"]
    response = requests.get(url)
    print(presi.get_text(), response.status_code)
    
    # parse & store html
    soup = BeautifulSoup(response.text, "lxml")
    presi_soups.append(soup.select(".infobox"))
    
    # respectful nap:
    wait_time = randint(1,2)
    print("I will sleep for " + str(wait_time) + " second/s.")
    #sleep(wait_time)
```

### Finding and storing information about each president.

```python
#Birthday
print(presi_soups[20][0].select(".bday")[0].get_text())

#Political party
print(presi_soups[20][0].find("th", string="Political party").parent.find("a").get_text())

#Number of sons/daughters
print(len(presi_soups[20][0].find("th", string="Children").parent.find_all("li")))
```

```python
#using loop
```

```python
children=[]
for i in range(len(presi_soups)):
    try:
        children.append(len(presi_soups[i][0].find("th", string="Children").parent.find_all("li")))
    except:
        children.append(0.0)
```

```python
birthday = []
for i in presi_soups:
    try:
        birthday.append(presi_soups[i][0].select(".bday")[0].get_text())
    except:
        birthday.append(0)
        
```

```python
politcal_party = []
for i in range(len(presi_soups)):
    try:
        politcal_party.append(presi_soups[i][0].find("th", string="Political party").parent.find("a").get_text())
    except:
        politcal_party.append('non')
```
